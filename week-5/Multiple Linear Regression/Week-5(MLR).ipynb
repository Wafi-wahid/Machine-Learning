```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Python function to read data from file

```python
def get_path(path):
  train=pd.read_csv(path)
  return train
```

# Reading data


```python
get_path("/content/drive/MyDrive/ML/CO2 Emissions_Canada (1).csv")
```

```python
fdata = pd.read_csv("/content/drive/MyDrive/ML/CO2 Emissions_Canada (1).csv")

# take a look at the dataset
fdata.head()
```

# Let's select some features that we want to use for regression.

```python
cfdata = fdata[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cfdata.head(9)
```

# Let's plot Emission values with respect to Engine size:

```python
plt.scatter(cfdata.ENGINESIZE, cfdata.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

# Creating train and test dataset

```python
msk = np.random.rand(len(fdata)) < 0.8
train = cfdata[msk]
test = cfdata[~msk]
```

# Train data distribution

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

# Multiple Regression Model

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

# Ordinary Least Squares (OLS)

```python
y_hat= regr.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
x = np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(test[['CO2EMISSIONS']])
print("Mean Squared Error (MSE) : %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
