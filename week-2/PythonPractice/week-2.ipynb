```python
weight= int(input("Input your weight: "))
unit= input("(L)bs or (K)g: ")
if unit.upper() == "L":
  converted = (weight) * 0.45
  print(f"You are {str(converted)} kilos")
  print("done")
else:
  converted = (weight) / 0.45
  print(f"You are {str(converted)} pounds")
  print("done")
```

```python
names=["Wafi", "Wahid", "Ahmad", "Azhar"]
for name in names:
  print(name)

print("Print 0 to 3")
print(names[0:3])

print("Print 1 onward")
print(names[1:])

print("Print backside last element in array")
print(names[-1])

print("Print last 2 elements of array")
print(names[-2:])

print("Remove wfi")
names.remove("Wafi")
print(names)

print("append wafi")
names.append("Wafi")
print(names)

print("insert wafi at 0 position in array")
names.insert(0, "Wafi")
print(names)

print("clear whole array")
names.clear()
print(names)
```
