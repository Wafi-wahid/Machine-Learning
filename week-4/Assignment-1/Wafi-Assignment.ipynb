# Convert 'normalized-losses' column to numeric,

```python
automobile_data['normalized-losses'] = pd.to_numeric(automobile_data['normalized-losses'], errors='coerce')
```

# Imutation

```python
automobile_data['normalized-losses'].mean()

automobile_data['normalized-losses'] = automobile_data['normalized-losses'].fillna(automobile_data['normalized-losses'].mean())
```

# To find null values if any

```python
automobile_data['normalized-losses'].isnull().sum()
```

# To remove null values not full price column

```python
automobile_cleaned=automobile_data.dropna(subset=['price'])
```
